Today I want to go through some of the most common steps when doing time series analysis.

I'll use Python and data about appliances and light energy cosumption. You can find the dataset [here](https://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

PROJECT_DIR = "../"
DATA_DIR = "data/"
filename = "energydata_complete.csv"

plt.style.use('ggplot')

plt.rcParams["grid.alpha"] = 0.9
plt.rcParams["axes.facecolor"] = "#f0f0f0"
plt.rcParams["figure.facecolor"] = "#f0f0f0"
plt.rcParams["figure.figsize"] = (6.4*1.2, 4.8*1.2)
```

Let's load the dataset and see how it looks like.

```python
df = pd.read_csv(PROJECT_DIR+DATA_DIR+filename)
df.head()
```

The dataset contains attributes related to humidity and temperature besides from energy appliance and light consumption.

In this occasion I will only use the energy consumption variables to perform a preliminar exploratory data analysis. Just to keep it simple and focus on the time series analysis principles.

```python
df = df[["date", "Appliances", "lights"]]

df.describe()
```

Let's see how python is interpreting the data.

```python
df.dtypes
```

A time series dataset has the peculiarity of having a date-time attribute, and that data is a function of time. As Python thinks that the `date` variable is of type `object` we will have to transform it to a `datetime` object.

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

Another standard with this type of data is to set the column that contains the time dimension as the dataframe index.

```python
df.set_index("date", inplace=True)
```

Now, let's see the period that is covered by the data.

```python
print("Dates goes from {} to {}".format(df.index.min(), df.index.max()))
```

Another usfull thing to do is to compute columns from each of the components of the time dimension: year, month, date, hour, etc. They usually are potential attributes for a time series model.

```python
df["month"] = [m.strftime("%b") for m in df.index]
df["day"] = [d.strftime("%a") for d in df.index]
df["hour_minute"] = [h.strftime("%H:%M") for h in df.index]
```

# Data exploration

```python
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.plot(df["Appliances"], label="Appliances")
plt.legend()
plt.subplot(212)
plt.plot(df["lights"], label="Lights")
plt.legend()
plt.show()
```

With scatterplots we will see patterns more clearly.

```python
axes = df[["Appliances", "lights"]].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('10 Min. Totals (GWh)')
```

```python
axes = df.loc["2016-01",["Appliances", "lights"]].plot(marker='.', linestyle=None, figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('10 Min. Totals (GWh)')
```

```python
hourly = df.groupby(pd.Grouper(freq = "H")).sum()
hourly["hour"] = hourly.index.hour
hourly.sort_index(inplace=True)
```

```python
axes = hourly.loc["2016-01",["Appliances", "lights"]].plot(marker='.', linestyle=None, figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Hourly Totals (kWh)')
```

```python
daily = df.groupby(pd.Grouper(freq = "D")).sum()
daily["day"] = daily.index.day
```

```python
axes = daily.loc["2016-01",["Appliances", "lights"]].plot(marker='.', linestyle=None, figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily Totals (kWh)')
```

```python
fig, (ax1, ax2, ax3, ax4) =plt.subplots(4,1)
ax1.plot(df.loc["2016-01-11":"2016-01-18",:].Appliances)
ax2.plot(df.loc["2016-01-18":"2016-01-25",:].Appliances)
ax3.plot(df.loc["2016-01-25":"2016-02-01",:].Appliances)
ax4.plot(df.loc["2016-02-01":"2016-02-08",:].Appliances)
plt.show()
```

# Frequencies

```python
df = df.asfreq('10min')
```

```python
df.isnull().sum()
```

# Resampling

```python
hourly_mean = df[["Appliances", "lights"]].resample('H').mean()
daily_mean = df[["Appliances", "lights"]].resample('D').mean()
```

```python
start, end = '2016-02-01', "2016-02-07"
# Plot 10 min. and hourly resampled time series together
fig, ax = plt.subplots()
ax.plot(df.loc[start:end, 'Appliances'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(hourly_mean.loc[start:end, 'Appliances'],
marker='o', markersize=8, linestyle='-', label='Hourly Mean Resample')
ax.set_ylabel('Appliance Consumption (kWh)')
ax.legend();
```
