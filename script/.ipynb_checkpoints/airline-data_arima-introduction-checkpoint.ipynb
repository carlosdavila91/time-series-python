## A quick summary of ARIMA

### ARIMA Model for TS Forecasting

ARIMA is specified by three order parameters: (p, d, q)

**AR(q) Autoregression** - An auto regressive component refers to the use of past values in the regression equation for the time series.

**I(d) Integration** - uses differencing of observations (subtracting an observation from the observation at the previous time step)

**MA(q) Moving Average** - A moving average component depicts the error of the model as a combination of previous error terms. The order _q_ represents the number of terms to be included in the model.

### Various flavors of ARIMA Model

**ARIMA**: Non-seasonal Autoregressive Integrated Moving Average
**SARIMA**: Seasonal ARIMA
**SARIMAX**: Seasonal ARIMA with exogenous variables

**Pyramid Auto-ARIMA**
Use `auto_arima` function form `pmdarima` to identify the most optimal parameters for an ARIMA model and returns a fitted ARIMA model.

It returns:

+ `p`: number of autoregressive terms (AR)
+ `q`: number of moving average terms (MA)
+ `d`: number of non-seasonal differences

Acknowledge:

+ `p`, `d` & `q` represent non-seasonal components
+ `P`, `D` & `Q` represent seasonal components

```python
from pmdarima.arima import auto_arima

arima_model = auto_arima(df["Passengers"], start_p=1, d=1, start_q=1,
                        max_p=5, max_q=5, max_d=5, m=12,
                        start_P=0, D=1, start_Q=0, max_P=5, max_D=5, max_Q=5,
                        seasonal=True,
                        trace=True,
                        error_action="ignore",
                        suppress_warinings=True,
                        stepwise=True, n_fits=50)

arima_model.summary()
```

**Best model: SARIMAX(0, 1, 1)x(2, 1, 1, 12)**

Now fit the model and test it.

```python
size = int(len(df)*.66)

X_train, X_test = df[0:size], df[size:len(df)]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX 
  
model = SARIMAX(X_train['Passengers'],  
                order = (0, 1, 1),  
                seasonal_order =(2, 1, 1, 12)) 
  
result = model.fit() 
result.summary() 
```

Train prediction.

The predictions are offset by 1!

```python
start_index = 0
end_index = len(X_train)-1
train_pred = result.predict(start_index, end_index)

pd.concat([X_train.Passengers.head(), train_pred.rename("pred").head()], axis=1)
```

Let's see how it performs over the test set.

```python
start_index = len(X_train)
end_index = len(df)-1
test_pred = result.predict(start_index, end_index).rename('Predicted passengers') 

pd.concat([X_test.Passengers.head(), test_pred.head()], axis=1)
```

```python
test_pred.plot(legend=True)
X_test["Passengers"].plot(legend=True)
plt.show()
```
