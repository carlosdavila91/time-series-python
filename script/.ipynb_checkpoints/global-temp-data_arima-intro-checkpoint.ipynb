```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_DIR = "../"

plt.style.use('ggplot')

plt.rcParams["grid.alpha"] = 0.9
plt.rcParams["axes.facecolor"] = "#f0f0f0"
plt.rcParams["figure.facecolor"] = "#f0f0f0"
plt.rcParams["figure.figsize"] = (6.4*1.2, 4.8*1.2)
```

```python
df = pd.read_csv(PROJECT_DIR+"data/GlobalTemperatures.csv")
```

```python
df.dtypes
```

```python
df["dt"] = pd.to_datetime(df["dt"])
df.dtypes
```

```python
df.set_index("dt", inplace=True)
```

```python
plt.plot(df["LandAverageTemperature"])
plt.show()
```

Count NAs.

```python
len(df) - df.count()
```

```python
df['LandAverageTemperature'].interpolate(method="linear", inplace=True)
df['LandAverageTemperatureUncertainty'].interpolate(method="linear", inplace=True)
```

```python
from statsmodels.tsa.stattools import adfuller

adf, pvalue, usedlag_, nobs_, critical_values_, icbest_ = adfuller(df["LandAverageTemperature"])

print("pvalue = ", pvalue, ". If above 0.05, data is not stationary")
```

```python
df["year"] = [d.year for d in df.index]
df["year"].unique()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposed = seasonal_decompose(df["LandAverageTemperature"], model="additive")
```

```python
trend = decomposed.trend
seasonal = decomposed.seasonal
residual = decomposed.resid
```

```python
plt.figure(figsize=(12,8))
plt.subplot(311)
plt.plot(df["LandAverageTemperature"], label="Original")
plt.legend(loc = "upper left")
plt.subplot(312)
plt.plot(trend, label="Trend")
plt.legend(loc = "upper left")
plt.subplot(313)
plt.plot(residual, label="Residuals")
plt.legend(loc = "upper left")
plt.show()
```
